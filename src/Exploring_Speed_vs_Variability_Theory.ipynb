```python
from scipy.stats import lognorm
dist_mean = lambda mu, sigma: lognorm(sigma, scale=exp(mu), loc=0).mean()
dist_entropy = lambda mu, sigma: lognorm(sigma, scale=exp(mu), loc=0).entropy()

mu = 2
sigmas = arange(100)
mus = ones(len(sigmas))*mu
plot(list(map(dist_mean, mus, sigmas)), list(map(dist_entropy, mus, sigmas)))
xscale('log')

title("Relationship between Mean and Entropy,\nVarying Sigma, Constant Mu")
xlabel("Mean")
ylabel("Entropy")
```
